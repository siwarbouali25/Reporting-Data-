```python
import pandas as pd
import numpy as np
import random
from typing import List, Dict, Any

random.seed(42)
np.random.seed(42)
```

```python
companies = [
    {"company_id": "C001", "company_name": "NorthBank Tunis", "sector": "Banking", "subsector": "Commercial Banks", "country": "Tunisia"},
    {"company_id": "C002", "company_name": "MedInsure", "sector": "Insurance", "subsector": "Non-life Insurance", "country": "Tunisia"},
    {"company_id": "C003", "company_name": "TransLog Africa", "sector": "Transportation", "subsector": "Logistics & Freight", "country": "Tunisia"},
    {"company_id": "C004", "company_name": "IndusCem", "sector": "Industrial", "subsector": "Construction Materials", "country": "Tunisia"},
]

years = list(range(2020, 2025))

company_years = pd.DataFrame(
    [
        {
            "company_id": c["company_id"],
            "company_name": c["company_name"],
            "sector": c["sector"],
            "subsector": c["subsector"],
            "country": c["country"],
            "reporting_year": y
        }
        for c in companies
        for y in years
    ]
)

company_years.head()
```

```python
def generate_energy_raw(row):
    sector = row["sector"]

    if sector == "Banking":
        diesel = np.random.uniform(2000, 12000)
        gas = np.random.uniform(500, 3000)
        electricity = np.random.uniform(2000, 12000)
    elif sector == "Insurance":
        diesel = np.random.uniform(1000, 8000)
        gas = np.random.uniform(300, 2000)
        electricity = np.random.uniform(1500, 9000)
    elif sector == "Transportation":
        diesel = np.random.uniform(80000, 400000)
        gas = np.random.uniform(1000, 5000)
        electricity = np.random.uniform(3000, 15000)
    else:  # Industrial
        diesel = np.random.uniform(30000, 150000)
        gas = np.random.uniform(20000, 100000)
        electricity = np.random.uniform(20000, 120000)

    return {
        "company_id": row["company_id"],
        "reporting_year": row["reporting_year"],
        "diesel_liters": round(diesel, 2),
        "natural_gas_mwh": round(gas, 2),
        "electricity_purchased_mwh": round(electricity, 2),
        "electricity_renewable_mwh": round(electricity * np.random.uniform(0.05, 0.4), 2),
        "refrigerant_leak_kg": round(np.random.uniform(5, 150), 2),
        "emission_factor_source": random.choice(["DEFRA", "EPA", "National Grid Database"]),
        "boundary_type": random.choice(["operational_control", "financial_control"]),
        "data_quality_level": random.choice(["high", "medium", "medium", "low"])
    }

energy_raw_df = pd.DataFrame([generate_energy_raw(r) for _, r in company_years.iterrows()])
energy_raw_df.head()
```

```python
energy_raw_df.to_csv("energy_raw_df.csv", index=False)
print("Saved: energy_raw_df.csv")
```

```python
def generate_scope3_raw(row):
    sector = row["sector"]

    if sector == "Banking":
        financed = np.random.uniform(100000, 800000)
        purchased_goods = np.random.uniform(500000, 3000000)
        business_travel = np.random.uniform(200000, 1000000)
    elif sector == "Insurance":
        financed = np.random.uniform(50000, 300000)
        purchased_goods = np.random.uniform(300000, 1500000)
        business_travel = np.random.uniform(100000, 500000)
    elif sector == "Transportation":
        financed = np.random.uniform(10000, 50000)
        purchased_goods = np.random.uniform(1000000, 5000000)
        business_travel = np.random.uniform(50000, 300000)
    else:
        financed = np.random.uniform(10000, 70000)
        purchased_goods = np.random.uniform(2000000, 8000000)
        business_travel = np.random.uniform(50000, 250000)

    return {
        "company_id": row["company_id"],
        "reporting_year": row["reporting_year"],
        "scope3_cat1_purchased_goods_spend_usd": round(purchased_goods, 2),
        "scope3_cat6_business_travel_km": round(business_travel, 2),
        "scope3_cat7_employee_commuting_km": round(np.random.uniform(100000, 800000), 2),
        "scope3_cat15_financed_activity_value": round(financed, 2),
        "scope3_method": random.choice(["spend-based", "activity-based", "supplier-specific", "hybrid"]),
        "scope3_data_quality_level": random.choice(["high", "medium", "medium", "low"])
    }

scope3_raw_df = pd.DataFrame([generate_scope3_raw(r) for _, r in company_years.iterrows()])
scope3_raw_df.head()
```

```python
scope3_raw_df.to_csv("scope3_raw_df.csv", index=False)
print("Saved: scope3_raw_df.csv")
```

```python
def generate_finance_raw(row):
    sector = row["sector"]

    if sector == "Banking":
        total_assets = np.random.uniform(5e8, 4e9)
        climate_risk_assets = total_assets * np.random.uniform(0.08, 0.35)
        capex = np.random.uniform(2e6, 2e7)
    elif sector == "Insurance":
        total_assets = np.random.uniform(2e8, 2e9)
        climate_risk_assets = total_assets * np.random.uniform(0.05, 0.25)
        capex = np.random.uniform(1e6, 8e6)
    elif sector == "Transportation":
        total_assets = np.random.uniform(5e7, 8e8)
        climate_risk_assets = total_assets * np.random.uniform(0.15, 0.45)
        capex = np.random.uniform(5e6, 5e7)
    else:
        total_assets = np.random.uniform(1e8, 1.5e9)
        climate_risk_assets = total_assets * np.random.uniform(0.2, 0.5)
        capex = np.random.uniform(1e7, 1e8)

    climate_capex = capex * np.random.uniform(0.05, 0.45)

    return {
        "company_id": row["company_id"],
        "reporting_year": row["reporting_year"],
        "total_assets_usd": round(total_assets, 2),
        "high_climate_risk_assets_usd": round(climate_risk_assets, 2),
        "total_capex_usd": round(capex, 2),
        "climate_related_capex_usd": round(climate_capex, 2),
        "current_climate_loss_usd": round(np.random.uniform(0, capex * 0.2), 2),
        "finance_data_source": random.choice(["ERP", "asset register", "finance consolidation file"])
    }

finance_raw_df = pd.DataFrame([generate_finance_raw(r) for _, r in company_years.iterrows()])
finance_raw_df.head()
```

```python
def generate_risk_register_raw(row):
    sector = row["sector"]

    risks = {
        "Banking": [
            "Carbon-intensive portfolio exposure",
            "Flood exposure of financed real estate",
            "Regulatory climate disclosure non-compliance"
        ],
        "Insurance": [
            "Increase in catastrophe claims",
            "Underpricing of climate-related underwriting risk",
            "Regulatory reporting gap"
        ],
        "Transportation": [
            "Fuel price and carbon cost exposure",
            "Fleet transition delay",
            "Physical disruption from extreme heat"
        ],
        "Industrial": [
            "Carbon-intensive production exposure",
            "Water stress affecting operations",
            "Transition cost from emissions regulation"
        ]
    }

    risk_desc = random.choice(risks[sector])
    likelihood = random.randint(1, 5)
    magnitude = random.randint(1, 5)

    return {
        "company_id": row["company_id"],
        "reporting_year": row["reporting_year"],
        "risk_description": risk_desc,
        "risk_type": random.choice(["physical", "transition"]),
        "likelihood_score": likelihood,
        "magnitude_score": magnitude,
        "time_horizon": random.choice(["short-term", "medium-term", "long-term"]),
        "prioritization_method": random.choice(["likelihood_x_magnitude", "expert judgement", "multi-factor scoring"]),
        "mitigation_plan_exists": random.choice([True, True, False]),
        "risk_owner_function": random.choice(["risk", "finance", "operations", "sustainability"])
    }

risk_raw_df = pd.DataFrame([generate_risk_register_raw(r) for _, r in company_years.iterrows()])
risk_raw_df.head()
```

```python
GHG_PROTOCOL_STANDARDS = [
    "GHG Protocol Corporate Standard",
    "GHG Protocol Corporate Accounting and Reporting Standard"
]

CONSOLIDATION_APPROACHES = [
    "operational_control",
    "financial_control",
    "equity_share"
]

SCOPE2_BASIS = [
    "location-based",
    "market-based",
    "both"
]

EF_SOURCES_SCOPE1 = [
    "DEFRA 2024",
    "EPA 2024",
    "IPCC 2006 Guidelines",
    "National emission factor database 2024"
]

EF_SOURCES_SCOPE2 = [
    "National grid factor 2024",
    "IEA electricity emission factors",
    "DEFRA electricity factors 2024",
    "Residual mix supplier factors"
]

EF_SOURCES_SCOPE3 = [
    "EPA EEIO",
    "EXIOBASE",
    "Supplier-specific data",
    "Hybrid method using spend-based and activity-based factors"
]

DATA_QUALITY_NOTES = [
    "high",
    "moderate",
    "mixed"
]

SOURCE_DOC_TYPES = [
    "internal methodology memo",
    "GHG inventory manual",
    "sustainability reporting methodology",
    "emissions calculation protocol"
]
```

```python
def generate_ghg_methodology_row(base_row: Dict[str, Any]) -> Dict[str, Any]:
    scope2_basis = pick_one(SCOPE2_BASIS)

    scope1_stationary = pick_yes_no(0.9)
    scope1_mobile = pick_yes_no(0.8)
    scope1_refrigerants = pick_yes_no(0.6)

    third_party_verified = pick_yes_no(0.55)

    row = {
        "company_id": base_row["company_id"],
        "reporting_year": base_row["reporting_year"],

        "ghg_protocol_standard": pick_one(GHG_PROTOCOL_STANDARDS),
        "ghg_protocol_version": "revised edition",
        "consolidation_approach": pick_one(CONSOLIDATION_APPROACHES),

        "organizational_boundary_note": "Covers entities included under the selected consolidation approach.",
        "reporting_boundary_note": "Reporting boundary includes material operations and excludes immaterial activities where relevant.",

        "scope1_methodology": "Activity data multiplied by fuel, combustion and refrigerant emission factors.",
        "scope2_methodology": f"Purchased electricity emissions calculated using {scope2_basis} approach.",
        "scope3_methodology": "Scope 3 estimated using a mix of spend-based, activity-based and supplier-specific methods depending on data availability.",

        "scope2_basis_used": scope2_basis,
        "scope1_includes_refrigerants": scope1_refrigerants,
        "scope1_includes_stationary_combustion": scope1_stationary,
        "scope1_includes_mobile_combustion": scope1_mobile,

        "emission_factor_source_scope1": pick_one(EF_SOURCES_SCOPE1),
        "emission_factor_source_scope2": pick_one(EF_SOURCES_SCOPE2),
        "emission_factor_source_scope3": pick_one(EF_SOURCES_SCOPE3),

        "base_year": max(2020, int(base_row["reporting_year"]) - 2),
        "base_year_recalculation_policy": "Base year is recalculated when structural or methodological changes materially affect comparability.",
        "material_exclusions_note": "Non-material sources may be excluded where data is unavailable or immaterial.",
        "estimation_uncertainty_note": "Certain values rely on estimates, proxies, and supplier assumptions where primary data is unavailable.",
        "data_quality_note": pick_one(DATA_QUALITY_NOTES),

        "third_party_verified": third_party_verified,
        "verification_note": (
            "Limited assurance obtained for selected GHG indicators."
            if third_party_verified else
            "No external assurance obtained for methodology disclosures."
        ),

        "source_document_name": f"GHG Methodology Memo {base_row['reporting_year']}",
        "source_document_type": pick_one(SOURCE_DOC_TYPES)
    }

    # Optional passthrough if these columns exist
    for col in ["sector", "subsector", "country", "region"]:
        if col in base_row:
            row[col] = base_row[col]

    return row
```

```python
BOARD_OVERSIGHT_BODIES = [
    "Board of Directors",
    "Board Sustainability Committee",
    "Board Risk Committee",
    "Board Audit Committee"
]

MANAGEMENT_OVERSIGHT_BODIES = [
    "Executive Committee",
    "Chief Sustainability Officer",
    "Chief Risk Officer",
    "ESG Steering Committee",
    "Sustainability Department"
]

REVIEW_FREQUENCIES = [
    "monthly",
    "quarterly",
    "semi-annually",
    "annually"
]

COMPETENCY_APPROACHES = [
    "board training sessions",
    "external expert briefings",
    "internal sustainability workshops",
    "cross-functional governance updates"
]

REMUNERATION_LINK_TYPES = [
    "climate targets",
    "emissions reduction objectives",
    "broader ESG performance",
    "safety and sustainability KPIs"
]

SOURCE_DOC_TYPES_GOV = [
    "board charter",
    "committee terms of reference",
    "governance policy",
    "remuneration policy",
    "annual governance memo"
]
```

```python
def generate_governance_row(base_row: Dict[str, Any]) -> Dict[str, Any]:
    board_oversight = random.random() < 0.85
    management_oversight = random.random() < 0.92
    remuneration_linked = random.random() < 0.45
    formal_responsibility = random.random() < 0.88
    competency_process = random.random() < 0.72

    board_body = pick_one(BOARD_OVERSIGHT_BODIES) if board_oversight else None
    management_body = pick_one(MANAGEMENT_OVERSIGHT_BODIES) if management_oversight else None

    row = {
        "company_id": base_row["company_id"],
        "reporting_year": base_row["reporting_year"],

        "board_oversight_exists": board_oversight,
        "board_oversight_body": board_body,
        "board_review_frequency": pick_one(REVIEW_FREQUENCIES) if board_oversight else None,
        "board_oversight_note": (
            f"{board_body} reviews climate and sustainability-related matters."
            if board_oversight else
            "No formal board-level oversight identified."
        ),

        "management_oversight_exists": management_oversight,
        "management_oversight_body": management_body,
        "management_review_frequency": pick_one(REVIEW_FREQUENCIES) if management_oversight else None,
        "management_oversight_note": (
            f"{management_body} is responsible for day-to-day oversight of ESG and climate issues."
            if management_oversight else
            "No formal management-level oversight identified."
        ),

        "formal_roles_documented": formal_responsibility,
        "formal_roles_note": (
            "Roles and responsibilities are documented in governance policies or committee charters."
            if formal_responsibility else
            "Roles and responsibilities are not formally documented."
        ),

        "board_competency_process_exists": competency_process,
        "board_competency_approach": pick_one(COMPETENCY_APPROACHES) if competency_process else None,
        "board_competency_note": (
            "Board competency on sustainability-related issues is maintained through structured updates and training."
            if competency_process else
            "No formal competency-building process identified."
        ),

        "executive_remuneration_linked_to_esg": remuneration_linked,
        "executive_remuneration_link_type": pick_one(REMUNERATION_LINK_TYPES) if remuneration_linked else None,
        "executive_remuneration_note": (
            "Executive remuneration includes selected ESG- or climate-related performance indicators."
            if remuneration_linked else
            "No explicit ESG or climate linkage identified in executive remuneration."
        ),

        "governance_escalation_process_exists": random.random() < 0.7,
        "governance_integration_with_risk_exists": random.random() < 0.75,

        "source_document_name": f"Governance Source Pack {base_row['reporting_year']}",
        "source_document_type": pick_one(SOURCE_DOC_TYPES_GOV)
    }

    # derive notes for the two extra flags
    row["governance_escalation_process_note"] = (
        "Material sustainability matters can be escalated from management to board-level governance bodies."
        if row["governance_escalation_process_exists"] else
        "No formal escalation process identified."
    )

    row["governance_risk_integration_note"] = (
        "Governance bodies review sustainability matters in coordination with enterprise risk management processes."
        if row["governance_integration_with_risk_exists"] else
        "Governance and risk management processes appear only partially connected."
    )

    for col in ["sector", "subsector", "country", "region"]:
        if col in base_row:
            row[col] = base_row[col]

    return row
```

```python
TARGET_TYPES = [
    "absolute emissions reduction",
    "emissions intensity reduction",
    "renewable electricity target",
    "energy efficiency target",
    "net zero target"
]

TARGET_SCOPES = [
    "Scope 1",
    "Scope 2",
    "Scope 1 and 2",
    "Scope 1, 2 and 3"
]

TARGET_STATUSES = [
    "on track",
    "partially on track",
    "delayed",
    "not assessed"
]

VALIDATION_STATUSES = [
    "SBTi validated",
    "internally approved",
    "under review",
    "not validated"
]

TRANSITION_PLAN_STATUSES = [
    "formal published plan",
    "board-approved internal plan",
    "partial transition actions only",
    "no formal transition plan"
]

CARBON_CREDIT_TYPES = [
    "nature-based credits",
    "removal credits",
    "avoidance credits",
    "mixed portfolio"
]

CARBON_CREDIT_USE = [
    "residual emissions only",
    "interim target support",
    "voluntary offsetting only"
]

SOURCE_DOC_TYPES_TARGETS = [
    "climate transition plan",
    "target approval memo",
    "sustainability strategy",
    "board-approved decarbonization roadmap",
    "net zero commitment document"
]
```

```python
def generate_target_transition_row(base_row):
    has_target = random.random() < 0.85
    has_transition_plan = random.random() < 0.70
    uses_carbon_credits = random.random() < 0.35

    if has_target:
        target_type = pick_one(TARGET_TYPES)
        target_scope = pick_one(TARGET_SCOPES)
        base_year = random.choice([2019, 2020, 2021, 2022, 2023])
        target_year = random.choice([2030, 2035, 2040, 2045, 2050])

        # keep target year after reporting year
        if target_year <= base_row["reporting_year"]:
            target_year = base_row["reporting_year"] + 5

        interim_target_year = random.choice([2025, 2026, 2027, 2028, 2030])
        if interim_target_year >= target_year:
            interim_target_year = max(base_year + 3, target_year - 5)

        target_value_pct = random.choice([20, 25, 30, 40, 45, 50, 60, 75, 90])
        interim_target_value_pct = max(10, int(target_value_pct * random.uniform(0.3, 0.7)))

        target_status = pick_one(TARGET_STATUSES)
        validation_status = pick_one(VALIDATION_STATUSES)
    else:
        target_type = None
        target_scope = None
        base_year = None
        target_year = None
        interim_target_year = None
        target_value_pct = None
        interim_target_value_pct = None
        target_status = None
        validation_status = None

    if has_transition_plan:
        transition_plan_status = pick_one(TRANSITION_PLAN_STATUSES[:3])
        plan_covers_operations = random.random() < 0.85
        plan_covers_supply_chain = random.random() < 0.60
        plan_covers_capex_alignment = random.random() < 0.70
    else:
        transition_plan_status = "no formal transition plan"
        plan_covers_operations = False
        plan_covers_supply_chain = False
        plan_covers_capex_alignment = False

    if uses_carbon_credits:
        carbon_credit_type = pick_one(CARBON_CREDIT_TYPES)
        carbon_credit_planned_use = pick_one(CARBON_CREDIT_USE)
    else:
        carbon_credit_type = None
        carbon_credit_planned_use = None

    row = {
        "company_id": base_row["company_id"],
        "reporting_year": base_row["reporting_year"],

        "has_climate_target": has_target,
        "target_type": target_type,
        "target_scope": target_scope,
        "target_base_year": base_year,
        "target_year": target_year,
        "target_value_pct": target_value_pct,
        "interim_target_year": interim_target_year,
        "interim_target_value_pct": interim_target_value_pct,
        "target_status": target_status,
        "target_validation_status": validation_status,

        "has_transition_plan": has_transition_plan,
        "transition_plan_status": transition_plan_status,
        "transition_plan_covers_operations": plan_covers_operations,
        "transition_plan_covers_supply_chain": plan_covers_supply_chain,
        "transition_plan_covers_capex_alignment": plan_covers_capex_alignment,

        "uses_carbon_credits": uses_carbon_credits,
        "carbon_credit_type": carbon_credit_type,
        "carbon_credit_planned_use": carbon_credit_planned_use,

        "target_note": (
            f"The company has a {target_type} covering {target_scope} emissions."
            if has_target else
            "No formal climate target identified."
        ),

        "transition_plan_note": (
            "The company has a transition plan describing decarbonization actions and implementation priorities."
            if has_transition_plan else
            "No formal transition plan identified."
        ),

        "source_document_name": f"Target and Transition Pack {base_row['reporting_year']}",
        "source_document_type": pick_one(SOURCE_DOC_TYPES_TARGETS)
    }

    for col in ["sector", "subsector", "country", "region"]:
        if col in base_row:
            row[col] = base_row[col]

    return row
```

```python
def generate_scenario_analysis_row(base_row):
    scenario_used = random.random() < 0.65

    if scenario_used:
        scenario_1_name = pick_one(SCENARIO_NAMES)
        scenario_2_name = pick_one([s for s in SCENARIO_NAMES if s != scenario_1_name])

        scenario_1_temp = pick_one(TEMPERATURE_PATHWAYS[:3])
        scenario_2_temp = pick_one(TEMPERATURE_PATHWAYS[2:])

        short_term_horizon = pick_one(TIME_HORIZONS_SHORT)
        medium_term_horizon = pick_one(TIME_HORIZONS_MEDIUM)
        long_term_horizon = pick_one(TIME_HORIZONS_LONG)

        business_scope = pick_one(BUSINESS_SCOPE_OPTIONS)
        source_type = pick_one(ANALYSIS_SOURCE_TYPES)
        resilience_outcome = pick_one(RESILIENCE_OUTCOMES)

        assumptions = random.sample(ASSUMPTION_THEMES, k=3)
        assumptions_text = "; ".join(assumptions)

        row = {
            "company_id": base_row["company_id"],

            "reporting_year": base_row["reporting_year"],

            "scenario_analysis_used": True,
            "scenario_1_name": scenario_1_name,
            "scenario_2_name": scenario_2_name,
            "scenario_1_temperature_pathway": scenario_1_temp,
            "scenario_2_temperature_pathway": scenario_2_temp,

            "short_term_horizon": short_term_horizon,
            "medium_term_horizon": medium_term_horizon,
            "long_term_horizon": long_term_horizon,

            "business_scope_covered": business_scope,
            "analysis_source_type": source_type,
            "key_assumptions_summary": assumptions_text,
            "resilience_assessment_conclusion": resilience_outcome,

            "scenario_analysis_note": (
                f"Scenario analysis considered {scenario_1_name} and {scenario_2_name} across short-, medium-, and long-term horizons."
            ),

            "source_document_name": f"Scenario Analysis Pack {base_row['reporting_year']}",
            "source_document_type": pick_one(SOURCE_DOC_TYPES_SCENARIO)
        }
    else:
        row = {
            "company_id": base_row["company_id"],
            "reporting_year": base_row["reporting_year"],

            "scenario_analysis_used": False,
            "scenario_1_name": None,
            "scenario_2_name": None,
            "scenario_1_temperature_pathway": None,
            "scenario_2_temperature_pathway": None,

            "short_term_horizon": None,
            "medium_term_horizon": None,
            "long_term_horizon": None,

            "business_scope_covered": None,
            "analysis_source_type": None,
            "key_assumptions_summary": None,
            "resilience_assessment_conclusion": None,

            "scenario_analysis_note": "No formal scenario analysis identified for the reporting year.",

            "source_document_name": f"Scenario Analysis Pack {base_row['reporting_year']}",
            "source_document_type": pick_one(SOURCE_DOC_TYPES_SCENARIO)
        }

    for col in ["sector", "subsector", "country", "region"]:
        if col in base_row:
            row[col] = base_row[col]

    return row
```

Emissions calculations

```python
EMISSION_FACTORS = {
    "diesel_liters_to_tco2e": 0.00268,          # tCO2e per liter
    "natural_gas_mwh_to_tco2e": 0.202,          # tCO2e per MWh
    "refrigerant_kg_to_tco2e": 0.0015,          # simplified placeholder
    "grid_electricity_mwh_to_tco2e": 0.42       # tCO2e per MWh
}
```

```python
def calculate_scope1_scope2(energy_df: pd.DataFrame) -> pd.DataFrame:
    df = energy_df.copy()

    # Scope 1
    df["scope1_diesel_tco2e"] = df["diesel_liters"] * EMISSION_FACTORS["diesel_liters_to_tco2e"]
    df["scope1_natural_gas_tco2e"] = df["natural_gas_mwh"] * EMISSION_FACTORS["natural_gas_mwh_to_tco2e"]
    df["scope1_refrigerants_tco2e"] = df["refrigerant_leak_kg"] * EMISSION_FACTORS["refrigerant_kg_to_tco2e"]

    df["scope1_emissions_tco2e"] = (
        df["scope1_diesel_tco2e"] +
        df["scope1_natural_gas_tco2e"] +
        df["scope1_refrigerants_tco2e"]
    )

    # Scope 2 location-based
    df["scope2_location_based_tco2e"] = (
        df["electricity_purchased_mwh"] * EMISSION_FACTORS["grid_electricity_mwh_to_tco2e"]
    )

    # Scope 2 market-based
    residual_mwh = df["electricity_purchased_mwh"] - df["electricity_renewable_mwh"]
    residual_mwh = residual_mwh.clip(lower=0)

    df["scope2_market_based_tco2e"] = (
        residual_mwh * EMISSION_FACTORS["grid_electricity_mwh_to_tco2e"]
    )

    # Total energy
    df["total_energy_consumption_mwh"] = (
        df["natural_gas_mwh"] + df["electricity_purchased_mwh"]
    )

    # Methodology / disclosure-ready fields
    df["ghg_measurement_methodology"] = (
        "Scope 1 and Scope 2 calculated from activity data using emission factors."
    )
    df["emission_factor_source_disclosure"] = df["emission_factor_source"]
    df["organizational_boundary_disclosure"] = df["boundary_type"]

    keep_cols = [
        "company_id", "reporting_year",
        "scope1_diesel_tco2e", "scope1_natural_gas_tco2e", "scope1_refrigerants_tco2e",
        "scope1_emissions_tco2e",
        "scope2_location_based_tco2e", "scope2_market_based_tco2e",
        "total_energy_consumption_mwh",
        "ghg_measurement_methodology",
        "emission_factor_source_disclosure",
        "organizational_boundary_disclosure",
        "data_quality_level"
    ]
    return df[keep_cols]
```

```python
emissions_df = calculate_scope1_scope2(energy_raw_df)
emissions_df.head()
```

```python
SCOPE3_FACTORS = {
    "cat1_spend_to_tco2e": 0.00035,         # purchased goods per USD
    "cat6_business_travel_km_to_tco2e": 0.00018,
    "cat7_commuting_km_to_tco2e": 0.00009,
    "cat15_financed_to_tco2e": 0.0008
}
```

```python
def calculate_scope3(scope3_df: pd.DataFrame) -> pd.DataFrame:
    df = scope3_df.copy()

    df["scope3_cat1_tco2e"] = (
        df["scope3_cat1_purchased_goods_spend_usd"] * SCOPE3_FACTORS["cat1_spend_to_tco2e"]
    )

    df["scope3_cat6_tco2e"] = (
        df["scope3_cat6_business_travel_km"] * SCOPE3_FACTORS["cat6_business_travel_km_to_tco2e"]
    )

    df["scope3_cat7_tco2e"] = (
        df["scope3_cat7_employee_commuting_km"] * SCOPE3_FACTORS["cat7_commuting_km_to_tco2e"]
    )

    df["scope3_cat15_tco2e"] = (
        df["scope3_cat15_financed_activity_value"] * SCOPE3_FACTORS["cat15_financed_to_tco2e"]
    )

    df["scope3_total_tco2e"] = (
        df["scope3_cat1_tco2e"] +
        df["scope3_cat6_tco2e"] +
        df["scope3_cat7_tco2e"] +
        df["scope3_cat15_tco2e"]
    )

    df["scope3_methodology_note"] = (
        "Scope 3 estimated using a mix of spend-based, activity-based and proxy methods depending on data availability."
    )

    df["scope3_categories_disclosed"] = "1, 6, 7, 15"

    keep_cols = [
        "company_id", "reporting_year",
        "scope3_cat1_tco2e",
        "scope3_cat6_tco2e",
        "scope3_cat7_tco2e",
        "scope3_cat15_tco2e",
        "scope3_total_tco2e",
        "scope3_methodology_note",
        "scope3_categories_disclosed",
        "scope3_method",
        "scope3_data_quality_level"
    ]
    return df[keep_cols]
```

```python
def calculate_climate_finance_metrics(finance_df: pd.DataFrame) -> pd.DataFrame:
    df = finance_df.copy()

    df["climate_capex_pct"] = np.where(
        df["total_capex_usd"] > 0,
        (df["climate_related_capex_usd"] / df["total_capex_usd"]) * 100,
        np.nan
    )

    df["high_climate_risk_assets_pct"] = np.where(
        df["total_assets_usd"] > 0,
        (df["high_climate_risk_assets_usd"] / df["total_assets_usd"]) * 100,
        np.nan
    )

    df["financial_effect_current_year_level"] = pd.cut(
        df["current_climate_loss_usd"] / df["total_assets_usd"],
        bins=[-np.inf, 0.005, 0.02, np.inf],
        labels=["low", "medium", "high"]
    )

    df["climate_financial_disclosure_note"] = (
        "Climate-related financial metrics derived from asset exposure, climate-tagged capital expenditure and current climate-related losses."
    )

    keep_cols = [
        "company_id", "reporting_year",
        "total_assets_usd",
        "high_climate_risk_assets_usd",
        "total_capex_usd",
        "climate_related_capex_usd",
        "current_climate_loss_usd",
        "climate_capex_pct",
        "high_climate_risk_assets_pct",
        "financial_effect_current_year_level",
        "climate_financial_disclosure_note",
        "finance_data_source"
    ]
    return df[keep_cols]
```

```python
def calculate_climate_risk_metrics(risk_df: pd.DataFrame) -> pd.DataFrame:
    df = risk_df.copy()

    df["climate_risk_score"] = df["likelihood_score"] * df["magnitude_score"]

    df["climate_risk_level"] = pd.cut(
        df["climate_risk_score"],
        bins=[-np.inf, 6, 12, np.inf],
        labels=["low", "medium", "high"]
    )

    df["risk_assessment_methodology"] = (
        "Climate risks assessed using likelihood and magnitude scoring from the enterprise risk register."
    )

    df["risk_prioritization_basis"] = df["prioritization_method"]

    df["risk_assessment_structured_disclosure"] = (
        "Likelihood and magnitude are used to assess climate-related risks and opportunities."
    )

    keep_cols = [
        "company_id", "reporting_year",
        "risk_description",
        "risk_type",
        "likelihood_score",
        "magnitude_score",
        "climate_risk_score",
        "climate_risk_level",
        "time_horizon",
        "mitigation_plan_exists",
        "risk_owner_function",
        "risk_assessment_methodology",
        "risk_prioritization_basis",
        "risk_assessment_structured_disclosure"
    ]
    return df[keep_cols]
```

```python
def classify_frequency(freq: int) -> str:
    if pd.isna(freq):
        return "unknown"
    if freq >= 12:
        return "monthly_or_more"
    elif freq >= 4:
        return "quarterly"
    elif freq >= 2:
        return "semi_annual"
    elif freq >= 1:
        return "annual"
    return "none"


def classify_governance_maturity(score: int) -> str:
    if score >= 8:
        return "high"
    elif score >= 5:
        return "medium"
    return "low"
```

```python
def calculate_governance_metrics(governance_df: pd.DataFrame) -> pd.DataFrame:
    df = governance_df.copy()

    # Core flags
    df["board_oversight_exists"] = df["board_reviews_climate"]
    df["committee_oversight_exists"] = (
        df["board_has_sustainability_committee"] & df["committee_reports_to_board"]
    )
    df["management_oversight_exists"] = df["management_climate_role_exists"]
    df["remuneration_linked_to_climate"] = df["climate_kpis_in_executive_remuneration"]

    # Frequency labels
    df["board_oversight_frequency_label"] = df["board_review_frequency_per_year"].apply(classify_frequency)
    df["internal_reporting_frequency_label"] = df["internal_reporting_frequency_per_year"].apply(classify_frequency)

    # Governance score
    score_cols = [
        "board_reviews_climate",
        "board_has_sustainability_committee",
        "committee_reports_to_board",
        "management_climate_role_exists",
        "climate_kpis_in_executive_remuneration",
        "board_climate_training_provided",
        "governance_policy_formalized",
        "climate_escalation_process_exists",
        "external_assurance_on_climate_disclosures"
    ]

    df["governance_score"] = df[score_cols].fillna(False).astype(int).sum(axis=1)
    df["governance_maturity"] = df["governance_score"].apply(classify_governance_maturity)

    # Structured disclosure fields
    df["governance_oversight_disclosure"] = np.where(
        df["board_oversight_exists"],
        "The board oversees climate-related matters.",
        "Climate-related matters are not formally overseen by the board."
    )

    df["management_role_disclosure"] = np.where(
        df["management_oversight_exists"],
        "Management has assigned responsibility for climate-related matters.",
        "Management responsibility for climate-related matters is not formally assigned."
    )

    df["remuneration_disclosure"] = np.where(
        df["remuneration_linked_to_climate"],
        "Executive remuneration includes climate-related performance considerations.",
        "Executive remuneration is not linked to climate-related performance."
    )

    df["committee_disclosure"] = np.where(
        df["committee_oversight_exists"],
        "A committee supports board oversight of climate-related matters.",
        "No formal committee structure supports climate-related oversight."
    )

    df["skills_and_competencies_disclosure"] = np.where(
        df["board_climate_training_provided"],
        "Climate-related training or capability development is provided to the board.",
        "No formal board climate-related training was identified."
    )

    keep_cols = [
        "company_id",
        "reporting_year",
        "board_oversight_exists",
        "committee_oversight_exists",
        "committee_name",
        "management_oversight_exists",
        "management_role_title",
        "remuneration_linked_to_climate",
        "board_oversight_frequency_label",
        "internal_reporting_frequency_label",
        "external_assurance_on_climate_disclosures",
        "governance_score",
        "governance_maturity",
        "governance_oversight_disclosure",
        "committee_disclosure",
        "management_role_disclosure",
        "remuneration_disclosure",
        "skills_and_competencies_disclosure"
    ]

    return df[keep_cols]
```
